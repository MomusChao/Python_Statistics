```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
 
# create data
x = np.random.rand(10)
y = x+np.random.rand(10)
z = x+np.random.rand(10)
z=z*z
 
plt.scatter(x, y, s=z*3000, c=x, cmap="Blues", alpha=0.4, edgecolors="grey", linewidth=2)

plt.xlabel("Variable 1")
plt.ylabel("Variable 2")
plt.title("Bubble Plot")
 
plt.show()
```
